```python
file = open('/Users/doyoung/학교/Python_Programming_JupyterNotebook/데이터/input.txt', 'r', encoding='utf-8')
data = file.readlines()
file.close()
char_list = data[0].split(' ') # 공백을 기준으로 데이터를 자름 -> 결과는 문자형으로 저장

data_list = []
for ch in char_list: # 문자형으로 받은 데이터를 정수형으로 변환하여 data_list에 대입
    data_list.append(int(ch))

capacity = 3 # 캐시 용량 설정
count = [0 for i in range(capacity)] # 설정한 캐시 용량만큼 count 블럭 생성
cache = [] # 캐시 블럭을 나타내는 변수
hit = 0 # hit 횟수를 count 하는 변수

for num in data_list: # 받은 데이터를 num 변수로 하나씩 가져옴
    if num == 99: # num 값이 99면 종료
        break
    else:
        if num in cache: # cache 안에 num이 존재한다면 hit
            hit += 1
            print("num : {} \t count : {} \t cache : {}".format(num, count, cache))
        else:
            if len(cache) == capacity: # cache 블럭에 값이 모두 찼을 경우 
                max_count = count.index(max(count)) # count값이 가장 높은 블럭 인덱스 값을 max_count에 저장
                del cache[max_count] # count 값이 가장 높은 캐시 블럭 값을 제거
                cache.insert(max_count, num) # 위에서 제거한 블럭에 num 값 대입
                count[max_count] = 0 # 들어간 캐시 블럭의 count 값을 0으로 초기화
                for i in range(0, len(cache)): # count 값 1씩 증가
                    count[i] += 1
                print("num : {} \t count : {} \t cache : {}".format(num, count, cache))
            else:
                cache.append(num) # 캐시 블럭에 num 값 추가
                for i in range(0, len(cache)): # count 값 1씩 증가
                    count[i] += 1
                print("num : {} \t count : {} \t cache : {}".format(num, count, cache))

for i in range(0, len(cache)):
    print("{}번 캐시 블록 값 : {}".format(i+1, cache[i]))
print("적중률 : {} / {}".format(hit, len(data_list)-1))
```
