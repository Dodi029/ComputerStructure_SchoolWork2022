```python
## 파일 읽기 파트 ##
f = open('/Users/doyoung/학교/Python_Programming_JupyterNotebook/데이터/input.txt', 'r', encoding='utf-8')
reference = f.readlines()
f.close()
data = reference[0].split(' ') # 공백을 기준으로 데이터를 분할해 문자형으로 저장

data_list = []
for ch in data:
    data_list.append(int(ch)) # 문자형으로 저장한 데이터를 정수형으로 변환해 data_list에 저장

## 변수/리스트 생성 파트 ##
hit : int = 0                             # 총 hit 횟수를 기록하는 변수
cache = []                                # 캐시 메모리를 나타내는 변수
capacity = 3                              # 캐시 메모리 용량 설정
count = [0 for i in range(capacity)]     # 설정한 캐시 용량만큼 count 블럭 생성 - 캐시 메모리에 값이 저장된 기간을 기록
hit_list = [0 for i in range(capacity)]  # 설정한 캐시 용량만큼 hit_list 블럭 생성 - 캐시 메모리 값이 각각 참조된 횟수 기록
loop = 0 ### 디버그용 ###

## LFU 알고리즘 ##
for num in data_list: # 받은 데이터를 num 변수로 하나씩 가져옴
    loop += 1 ### 디버그용 ###
    if (num == 99):   # num 값이 99면 반복문 종료
        break;
        
    else:
        
        if num in cache:                    # 입력된 num 값이 캐시 메모리에 존재할 경우 
            hit += 1                        # 총 hit 횟수 1 증가
            hit_list[cache.index(num)] += 1 # 해당 num 값의 hit_list 값 1 증가
            for i in range (0, len(count)):
                    count[i] += 1           # 이후 전체 리스트의 count 값 1 증가
            print("{:>2} num : {} \t count : {} \t cache : {} (Hit!!)".format(loop, num, count, cache))
            
            
        else:                                    # 입력된 num 값이 캐시 메모리에 존재하지 않을 경우
            if len(cache) < capacity:            # 캐시 메모리의 현재 크기가 설정된 용량 미만이면
                cache.append(num)                # 캐시 메모리에 num 값 append
                for i in range(0,len(cache)):
                    count[i] += 1                # 이후 전체 리스트의 count값 1 증가
                print("{:>2} num : {} \t count : {} \t cache : {}".format(loop, num, count, cache))
                    
            else:                                                        # 캐시 메모리의 현재 크기가 설정된 용량 이상이면
                if not (hit_list.count(min(hit_list)) > 1):              # 참조된 횟수가 최소인 값이 하나만 존재할 경우
                        cache.pop(hit_list.index(min(hit_list)))         # 그 값을 삭제
                        cache.insert(hit_list.index(min(hit_list)), num) # 이후 그 자리에 num 값 삽입
                        count[hit_list.index(min(hit_list))] = 0         # 이후 해당 값의 list_index 값을 0으로 초기화
                        
                else:                                                            # 참조된 횟수가 최소인 값이 중복 존재할 경우
                    cg = 0                                                       # 그 값들 중 가장 먼저 저장된 값을 찾기 위한 변수
                    for i in range(0, len(hit_list)):
                        if ((hit_list[i] == min(hit_list)) and (count[i] > cg)):# 참조된 횟수가 최소인 값 중 가장 먼저 저장된 값 탐색
                            cg = count[i]
                        else:
                            continue
                    cache.pop(count.index(cg))                                   # 탐색 결과 나온 값을 삭제(FIFO)
                    cache.insert(count.index(cg), num)                           # 이후 그 자리에 num 값 삽입
                    count[count.index(cg)] = 0                                   # 이후 해당 값의 list_index 값을 0으로 초기화
                        
                for i in range(0,len(count)):        # num 값 삽입이 완료되었으면 전체 리스트의 count값 1 증가
                        count[i] += 1
                print("{:>2} num : {} \t count : {} \t cache : {}".format(loop, num, count, cache))
            
## 최종 캐시 메모리 상태 및 적중률 출력 ##
print("\n")
for i in range(0, len(cache)):
    print("{}번 캐시 블록 값 : {}".format(i+1, cache[i]))
print("\n\n적중률 : {} / {}".format(hit, len(data_list)-1))
```
